```python
def get_object_infos(api, ids_or_refs, workspace=None):
    query = [api.process_workspace_identifiers(x, workspace) for x in ids_or_refs]
    ref_data = api.ws_client.get_object_info3(
        {"objects": query})
    ref_infos = [KBaseObjectInfo.from_kbase_json(d) for d in ref_data['infos']]
    return ref_infos
ref_data = get_object_infos(kbase, genome_set.refs)
```
