```python
matrix = kbase.get_from_ws('66231/68/6')
matrix_data = kbase.get_object('66231/68/6', None)
```

```python
print(type(matrix))
print(type(matrix_data))
```

```python
attr = kbase.get_from_ws(matrix.row_attribute_mapping)
attr_data = kbase.get_object(matrix.row_attribute_mapping, None)
```

```python
assert attr.get_data()['attributes'] == attr_data['attributes']
assert attr.get_data()['instances'] == attr_data['instances']
assert attr.get_data()['ontology_mapping_method'] == attr_data['ontology_mapping_method']
assert str(attr.info) == matrix.row_attribute_mapping
```

```python
assert matrix.get_data()['data']['col_ids'] == matrix_data['data']['col_ids']
assert matrix.get_data()['data']['row_ids'] == matrix_data['data']['row_ids']
assert matrix.get_data()['data']['values'] == matrix_data['data']['values']
assert matrix.get_data()['scale'] == matrix_data['scale']
assert matrix.get_data()['row_attributemapping_ref'] == matrix_data['row_attributemapping_ref']
assert matrix.get_data()['row_mapping'] == matrix_data['row_mapping']
assert matrix.get_data()['biochemistry_ref'] == matrix_data['biochemistry_ref']
assert matrix.get_data()['description'] == matrix_data['description']

#assert matrix.get_data()['genome_ref'] == matrix_data['genome_ref']
#assert matrix.get_data()['feature_mapping'] == matrix_data['feature_mapping']
#assert matrix.get_data() == matrix_data
```
