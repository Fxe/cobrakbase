```python
import cobrakbase
kbase = cobrakbase.KBaseAPI()
```

```python
model = kbase.get_from_ws('GCF_000005845.2.RAST.GMM_Anaerobic_akg.gf', 153317)
media = kbase.get_from_ws('Carbon-D-Glucose', 153317)
fba = kbase.get_from_ws('glucose.fba', 153317)
```

```python
model.summary()
```

```python
model.medium = media
model.summary()
```

```python
solution = {}
for rxn_var in fba.FBAReactionVariables:
    v = rxn_var['value']
    rxn_id = rxn_var['modelreaction_ref'].split('/')[-1]
    solution[rxn_id] = v
for rxn_var in fba.FBABiomassVariables:
    v = rxn_var['value']
    rxn_id = rxn_var['biomass_ref'].split('/')[-1]
    solution[rxn_id] = v
```

```python
def get_net_flux(solution, model):
    net_flux = {}
    for rxn in model.reactions:
        v = solution.get(rxn.id, 0)
        for m, coef in rxn.metabolites.items():
            if m.id not in net_flux:
                net_flux[m.id] = 0
            net_flux[m.id] += v * coef
    return net_flux
def clean_zeros(net):
    return {k:v for k, v in net.items() if round(v, 4) != 0}
clean_zeros(get_net_flux(solution, model))
```
